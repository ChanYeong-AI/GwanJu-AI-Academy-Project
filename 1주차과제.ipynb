음성 인식 처리 기술은 2000년 초에도 있었습니다만 많은 소음과 부정확한 발음 속에서도 사람의 음성을 정확하게 문장으로 치환하는 기술을 대중이 쉽게 접할 수 있게된 것은 그리 오래 된 일이 아닙니다.

그래서 그런지 많은 인공지능 기술들 중에서 가장 대중적으로 익숙한 것이 음성 인식 기술인지도 모르겠습니다.

오늘날에 음성 인식 기술을 활용한 사례로써 가장 대중적인 서비스는 바로 '가상 AI 서비스'라고 할 수 있습니다.

국내 3대 통신사 뿐만 아니라 많은 대기업들이 각자 하나씩 가상 AI 서비스를 제공하는 'AI 스피커'를 개발 및 판매하고 있습니다. 특정한 말을 하면 그에 맞는 명령을 수행하여 사람들이 더 편하고 즐거운 생활을 보낼 수 있도록 도와줍니다.

꼭 가상 AI 서비스가 아니더라도 사람의 음성을 인식하여 분류하고 그에 맞게 처리하는 작업은 앞으로의 인공지능 기술의 기반이 될 것이라고 생각합니다.

### 세 번째 기술 분야 : 자율주행 자동차
공상과학을 배경으로한 대중 매체들 중 '미래의 도시'를 보여줄 때 빠지지 않는 요소가 하나 있습니다. 바로 자율주행 자동차입니다.

자동차는 오래전부터 사회 발전에 있어 없어서는 안되는 중요한 물건이자 신속하고 편리한 운송 수단이고 자신의 목숨을 책임지는 파트너이기도 했습니다. 성인이라면 대부분 운전 면허는 기본이었고 안정적인 수입이 있다면 차 한대는 모두 가지고 있습니다. 회사의 사장 혹은 간부라면 지위를 내세우기 위해서도 필요한 것이 바로 개인 자동차입니다.

때문에 자동차는 연구 개발 투자에 있어 타 분야보다 경제적 위험성이 적다고 할 수 있습니다. 누구에게나 안전하고 편리하고 멋진 자동차가 하나쯤 있다면 나쁠 것이 없을테니까요.

자율주행기술의 선두를 달리고 있는 기업 중 'Tesla'라는 미국의 기업이 있습니다. 혁신적이고 도전적인 태도로 유명한 사업가, '일론 머스크(Elon Musk)'가 창립한 기업 중 한 곳이죠. 전기자동차로 유명하기도한 이 기업은 창립자의 정신에 걸맞게 혁신적이고 도전적인 제품과 기술을 개발하고 투자합니다. 

테슬라는 무려 수십억 마일에 달하는 주행 데이터를 가지고 있습니다. 인공지능의 정확도는 데이터의 양에 달려있다해도 무방할 정도로 방대한 데이터는 중요합니다.

### 네 번째 기술 분야 : 언어

마지막으로 언어입니다. 사람의 말은 서로에게 자신의 의사를 전달하기 위한 수단으로써 사용합니다. 지금까지 인간의 대화 상대는 항상 인간 뿐이었습니다만 오래전부터 인간은 인간이 아닌 존재와의 소통을 꿈꿔왔습니다. 과연 인공지능과 진정한 의미의 소통이 가능할지 정말 궁금합니다.

언어 기술 분야의 서비스 중 가장 인상 깊은 것은 바로 챗봇입니다. 챗봇은 사람과 일정 수준 이상의 대화가 가능한 인공지능 서비스를 말합니다. 특정한 주제에 대해서만 가능한 경우가 있지만 범용적인 대화도 가능한 경우가 있습니다. 위에서 소개한 가상 AI 서비스 또한 음성을 문장으로 치환하여 어느 정도의 대화가 가능합니다. 하지만 우리가 기대하는 수준의 대화라고는 할 수 없습니다.

스캐터랩의 챗봇 서비스인 '핑퐁'은 이러한 기대를 충족시키려하는 시도 중 하나라고 할 수 있습니다. 사람과 자연스러운 대화, 이어지는 대화를 할 수 있는 놀라운 챗봇 서비스 중 하나입니다. 사람들간의 대화를 100억 건 가량 학습하여 보다 친근하고 자연스러운 대화를 이어나갈 수 있습니다. 실제로 카카오톡의 '드림이', 구글 어시스턴트인 '파이팅 루나', '그남자 허세중'이라는 챗봇은 모두 핑퐁이 기반이 되었습니다.

최근에는 핑퐁을 통해 '핑퐁 빌더'라는 일상대화 기술 프레임워크 또한 공개하여 자신이 원하는 챗봇을 자유롭게 만들어내는 서비스 또한 공개하였습니다.
